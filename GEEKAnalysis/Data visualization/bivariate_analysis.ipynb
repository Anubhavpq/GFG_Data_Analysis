```python
import pandas as pd
import numpy as np
import seaborn as sns
df = sns.load_dataset("titanic")
df
```

```python
df = df.dropna()
df
```

```python
df.reset_index()
```

```python
# 1. reset the index and drop the old one in a single step
df = df.reset_index(drop=True)        # or
# df.reset_index(drop=True, inplace=True)

# now df has a 0…n index and no “index” column at all
df
```

```python
sns.barplot(x ="class", y= "age", data = df)
```

```python
sns.barplot(x ="class", y= "age",hue = "sex", data = df)
```

```python
sns.violinplot(x = "sex", y = "fare", data = df)
```
